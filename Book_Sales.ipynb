```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("/Users/rakadeb/Desktop/Books_Data_Clean.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df[df["Publishing Year"]>1990]
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.dropna(subset ="Book Name",inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution Of Publishing Year")
plt.show
```

```python
df["genre"].value_counts().plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("Number Of Books ")
plt.title("Number of Books in Each Genre ")
plt.show()
```

```python
sns.boxplot(x = "genre", y="Book_ratings_count",data=df)
plt.xlabel("Genre")
plt.ylabel("Book Ratings Count")
plt.title("Box Plot Of Book ratings count for Each Genre")
plt.show()
```

```python
plt.scatter(df["sale price"],df["units sold"])
plt.xlabel("Sale Price")
plt.ylabel("Units Sold")
plt.title("Scatter plot of sale prive vs Units sold")
plt.show()
```

```python
total_gross_sales_by_authors = df.groupby("Author")["gross sales"].sum()
```

```python
total_gross_sales_by_authors.sort_values(ascending=True).head().plot(kind="bar")
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total Gross Sales for Each Author")
plt.show()
```
